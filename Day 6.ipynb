```python
# key is gonna be a name of a country
# value is gonna be how many shows/movies are available in that country
countries = {}

for item in list(netflix['country']):

  # if there is a , -> we have multiple countries where that show is available
  if type(item) == type(''):
    if ',' in item:
      # turn it into a list
      countries_list = item.split(', ')

      #iterate through each country in the list
      for country in countries_list:
        # if country is in the keys add 1
        if country in countries.keys():
          countries[country] += 1
        # create the key and make it equal to one
        else:
          countries[country] = 1
    
    else:
      if item in countries.keys():
        countries[item] += 1
      else:
        countries[item] = 1
```

```python
top_country = ''
top_count = 0

for country, count in countries.items():

  if count > top_count:
    top_count = count
    top_country = country
```

## HW: Analyze the US Movie Durations using the Netflix Dataframe

1. Narrow down the dataframe to only include movies that were released in only USA

```python
# Hint: Try using boolean indexing
usa_movies = pd.DataFrame(netflix[netflix['country'] == 'United States'])
usa_movies
```

2. Remove any row that doesn't have its duration in minutes

```python
# Hint: try checking if "mins" is in the row
# Turn the duration column into a list
# create a new list
# if "mins" in the value than append 1 to new_list
# otherwise append 0
# netflix["valid"] = new_list
# use boolean indexing 

duration_list = usa_movies.duration
new_list = []

for duration in duration_list:
  if "min" in duration:
    new_list.append(1)
  else:
    new_list.append(0)

usa_movies["valid"] = new_list
usa_movies = usa_movies[usa_movies["valid"] == 1]
```

```python
usa_movies
```

3. Cast the type of the duration column into an integer

```python
# Hint: remove the 'mins' with replace and try using .strip()
# turn duration column into a list
# use .replace("mins", "")  and .strip()
# The cast type the value into an integer
# create a column in the dataframe like netflix["mins"] = list

duration_list = list(usa_movies.duration)

for i in range(len(duration_list)):
  duration_list[i] = int(duration_list[i].replace("min", "").strip())
  
usa_movies["mins"] = duration_list
```

```python
usa_movies
```

4. Find the average run time of the movie

```python
# Hint: .describe() should do the trick
# Use it on the new column in the "mins" column created in question 3

usa_movies.mins.mean()
```
