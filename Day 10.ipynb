## Practice with the Iris Dataset

```python
from sklearn import datasets
import pandas as pd
```

```python
iris = datasets.load_iris()

# Store features matrix in X
features = pd.DataFrame(iris.data, columns=['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'])

#Store target vector in 
flowers = ['setosa' 'versicolor' 'virginica']
target = pd.DataFrame(iris.target, columns=['names'])
```

image.png

```python
features.head()
```

```python
features = features.drop("flower", axis=1)
```

```python
target.head()
```

```python
 mapped = {0 :'setosa', 
          1 : 'versicolor',
          2 : 'virginica'}
```

image.png

```python
all_data = features
```

```python
all_data["flower"] = list(target["names"])
```

```python
mapped = {0 :'setosa', 
          1 : 'versicolor',
          2 : 'virginica'}
```

```python
actual_name = [mapped[target] for target in list(target["names"])]
```

```python
all_data["flower"] = actual_name
```

```python
all_data
```

```python
import plotly.express as px
```

```python
all_data.columns
```

```python
fig = px.scatter(all_data, x='petal length (cm)',y='petal width (cm)', color="flower")

fig.show()
```

```python
# split data into training and testing

# for knn()
from sklearn.neighbors import KNeighborsClassifier
# for splitting up the data
from sklearn.model_selection import train_test_split

# split data into training and testing set
X_train, X_test, y_train, y_test = train_test_split(features, target, random_state = 201)
```

```python
X_train
```

```python
# create and run the model

# select a classifier
knn = KNeighborsClassifier(n_neighbors=1)
# Create a model by fitting the training data
knn.fit(X = X_train, y = y_train)
```

```python
# evaluate the accuracy
predicted = knn.predict(X = X_test)
```

```python
accuracy = knn.score(X_test, y_test)
print(f'Prediction accuracy on the test data: {accuracy * 100}')
```

```python
for neighbors in [7,17,21,22,23,24,25]:

  knn = KNeighborsClassifier(n_neighbors=neighbors)

  # Create a model by fitting the training data
  knn.fit(X = X_train, y = y_train)

  # evaluate the accuracy

  predicted = knn.predict(X = X_test)

  accuracy = knn.score(X_test, y_test)
  print(f'Prediction accuracy on the test data with {neighbors} neighbors: {accuracy * 100}')
```

## Machine Learning with Linear Regression

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
california = fetch_california_housing() 
```

```python
print(california.DESCR)
```

```python
california.data.shape
```

```python
california.target.shape
```

```python
california.feature_names
```

```python
df = pd.DataFrame(california.data, columns = california.feature_names)
df['Value'] = california.target
```

```python
df.head()
```

```python
df.describe()
```

```python
features = df.drop('Value', axis = 1)
```

```python
features.head()
```

## Linear Regression

image.png

```python
target = df['Value']
target[:5]
```

```python
fig = px.scatter(df, x="AveBedrms", y="Value")

fig.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# split data into training and testing
X_train, X_test, y_train, y_test = train_test_split(features, target, random_state = 3000)
```

```python
y_test.head()
```

```python
# select a regressor and create the model by fitting the training data
model = LinearRegression().fit(X = X_train, y = y_train)
```

```python
model.coef_
```

```python
intercept = model.intercept_
intercept
```

```python
'''
Enumerate: adds a counter to an iterable and returns it and 
           the corresponding sequence item in a tuple
'''

for i, name in enumerate(california.feature_names):
    print(f'{name : >10}: {model.coef_[i]}')
```

```python
predicted = model.predict(X_test)
```

```python
expected = y_test
```

```python
for p, e in zip(predicted[:5], expected[:5]):
    print(f'predicted: {p:.2f}, expected: {e:.2f}')
```

```python
results_df = pd.DataFrame(expected.values, columns = ['expected'])
```

```python
results_df['predicted'] = predicted
```

```python
results_df
```

## Residuals

image.png

```python
from sklearn.metrics import r2_score
r2_score(expected, predicted)

# we can explain about 60% of the outcome variable
```
