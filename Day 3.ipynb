# Simulating Dice

```python
import random
```

```python
random.randint(1,6)
```

```python
roll_dist = {}
```

```python
for i in range(1000000):
  dice1 = random.randint(1,6)
  dice2 = random.randint(1,6)

  roll_sum = dice1 + dice2

  if roll_sum in roll_dist.keys():
    roll_dist[roll_sum] = roll_dist[roll_sum] + 1
  else:
    roll_dist[roll_sum] = 1
```

# Generating Recommendations

```python
netflix = {
    'Jasmine' : ['Frozen', 'Aladin'],
    'Anna' : ['Lion King', 'Tangled'],
    'Repunzel' : ['Frozen', 'Lion King', 'Aladin'],
    'Mulan' : ['Aladin'],
}

# What movies can I recommend to Mulan?
```

```python
similar_users = []
rec_movies = []

for movie in netflix['Mulan']:
  for user, movies in netflix.items():
    if user != 'Mulan':
      if movie in movies:
        similar_users.append(user)
  for user in similar_users:
    for other_movie in netflix[user]:
      if other_movie != movie:
        if other_movie not in rec_movies:
          rec_movies.append(other_movie)

print(rec_movies)
```

# Dictionary Review Questions

```python
changes = {
    'AAPL' : 5.34,
    'TSLA' : 10.89,
    'MSFT' : -3.89,
    'W' : 5.13,
    'GME' : 400
}
```

1. How would I add AMC with 200 to the dictionary?

```python
changes['AMC'] = 200
```

2. How many companies are in the dictionary?

```python
num_companies = len(changes.keys())
print(num_companies)
```

3. What is the average of all the changes?

```python
sum = 0

for num in changes.values():
  sum += num

avg = sum/num_companies
print(avg)
```

4. Which company has the highest change?

```python
highest_change = -100
company = ""

for name, num in changes.items():
  if num > highest_change:
    company = name
    highest_change = num

print(company)
```
